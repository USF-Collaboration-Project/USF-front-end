```python
import pandas as pd
import os
import numpy as np
import csv
```

```python
df = pd.read_csv('google_ee_v2.csv')
df.head(10)
```

```python
df['tags'] = df['tags'].apply(lambda x: np.array(x))
type(df['tags'].iloc[0])
```

```python
df['tags'] = df['tags'].apply(lambda x: x.splitlines())
```

```python
type(df['tags'])
```

```python
def find_unique(col):
    tags = set(x for lst in col for x in lst)
    return tags    
```

```python
tagss = find_unique(df['tags'])
print(tagss)
```

```python
tags = list(map(lambda x: x.lstrip(), tagss))
print(tags)
```

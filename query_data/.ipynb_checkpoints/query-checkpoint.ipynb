```python
import pandas as pd
import os
import numpy as np
import csv
```

```python
df = pd.read_csv('google_ee_v2.csv')
df.head(10)
```

```python
df['tags'] = df['tags'].str.replace('\n','')
```

```python
df.head(10)
```

```python
tags = df['tags'].unique().tolist()
print(tags)
```

```python
unique_tags = []
for element in tags:
    if element not in unique_tags:
        unique_tags.append(element)
    
print(unique_tags)
```
